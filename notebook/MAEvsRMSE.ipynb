```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
plt.rcParams["font.family"] = "Hiragino sans"
```

RMSEは予測値に局所的に大きく外す値がある場合、MAEよりも誤差が大きくなる  
MAEは局所的に大きく外す値があっても、平均されるのでいい感じにならされる  
RMSEは局所的に大きく外す値を2乗した上で平均を取るので、(√をとるとはいえ)その分MAEより大きくなる。

```python
x = np.arange(0, 5)
# 正解
y = np.array([1, 2, 0, 1, 0])
# 予測
y_pred = np.array([0, 2, 2, 10, 0])
print(f"MAE：{mean_absolute_error(y, y_pred)}")
print(f"RMSE：{np.sqrt(mean_squared_error(y, y_pred))}")

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot(111)
ax.plot(x, y, marker='.', label="y_true")
ax.plot(x, y_pred,marker='.', label="y_pred")
plt.legend()
plt.grid()
plt.show()
plt.close()
```

予測全体で大きく外してる場合はそんなにMAEと変わらない  
局所的に大きく外すことに敏感

```python
x = np.arange(0, 5)
# 正解
y = np.array([1, 2, 0, 1, 0])
# 予測
y_pred = np.array([201, 202, 200, 201, 200])
print(f"MAE：{mean_absolute_error(y, y_pred)}")
print(f"RMSE：{np.sqrt(mean_squared_error(y, y_pred))}")

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot(111)
ax.plot(x, y, marker='.', label="y_true")
ax.plot(x, y_pred,marker='.', label="y_pred")
plt.legend()
plt.grid()
plt.show()
plt.close()
```
